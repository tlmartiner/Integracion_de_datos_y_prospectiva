```python
Crmax=0; Idi=0 #Indice de Integración-Cuál vamos a integrar. #Buscar la credibilidad Máxima.
arrays_LDA=[] #Se almacenan  los datos de la FINTECH y sucursales
nxl='/content/drive/MyDrive/SEMESTRE 9/Integración de datos y prospectiva/Bases de datos/3. DB_Fintechs (USD)_Integración.xlsx'

for k in [0,1,2,3]:
  XDB=pd.read_excel(nxl,sheet_name=k)
  XDB=XDB.dropna() #Eliminar celdas vacías
  XDB=XDB[['Monto (EAD)','Prob.Default (PD)','LGD']]
  XDB['LDA']=XDB['Monto (EAD)']*XDB['Prob.Default (PD)']*XDB['LGD']
  arrays_LDA.append(XDB['LDA'])

  XDo=arrays_LDA[0] #FINTECH son los datos base

  if k>0: #Evita que se compare con ella misma.
    XDe=arrays_LDA[k] #Son los datos externos de las sucursales
    fc,Cr=credibilidad(XDo,XDe)

    print(f"La credibilidad de la sucursal {k} es: {(1-Cr)*100}")

  #¿Cuál de las sucursales vamos a integrar?
  if (1-Cr)>=Crmax:
    Crmx=Cr
    Idi=k

  print("La base de datos a integrar es:",Idi)

  #Se integran las bases de datos
  LDAo=arrays_LDA[0] #Base de datos antes de la Integración
  LDAo.describe()
  print("El coeficiente de asimetría es:",skew(LDAo))

  LDA_Int=np.concatenate((LDAo,arrays_LDA[Idi]),axis=0)
  LDA_Int=pd.DataFrame(LDA_Int)
  print("Los datos que describen los datos externos:\n")
  display(LDA_Int.describe())

  LDA_Int=np.concatenate((LDAo,arrays_LDA[Idi]),axis=0)
  LDA_Int=pd.DataFrame(LDA_Int)
  print("Las métricas que describen los datos integrados:\n")
  display(LDA_Int.describe())
```

2. Se procede con la eliminación de la credibilidad

```python
def credibilidad(XDo,XDe):

  #1. Se calcula el valor esperado de la varianza
  NDo=len(XDo); NDe=len(XDe)
  EPV=(NDo*np.var(XDo)+NDe*np.var(XDe))/(NDo+NDe) #Valor esperado de la varianza

  #2. Valor Hipotetico de la Media (Es la media compartida)
  uh=(NDo*np.mean(XDo)+NDe*np.mean(XDe))/(NDo+NDe)

  #3. Varianza Hipótetica de la Media (Varianza del total de los Datos
  #con respecto a la media compartida)
  VHM=((NDo*np.mean(XDo)**2+NDe*np.mean(XDe)**2)/(NDo+NDe))-uh**2

  #4. Factor de Credibilidad
  fc=(EPV/VHM)
  print("El factor de credibilidad es:",fc)

  #5. La Credibilidad es
  Cr=NDo/(NDo+fc)
  print("La credibilidad de la FINTECH base es:",(1-Cr)*100)

  return fc,Cr
```
